## Implement null filling

### Subtask:
Generate and execute the code to fill the null values in the `df_subset` DataFrame based on the chosen strategy.


**Reasoning**:
Drop the 'reviews.userCity' column and fill the null values in the specified columns with placeholder strings as determined in the previous step.



```python
df_subset = df_subset.drop(columns=['reviews.userCity'])
df_subset['reviews.title'] = df_subset['reviews.title'].fillna('Unknown Title')
df_subset['manufacturer'] = df_subset['manufacturer'].fillna('Unknown Manufacturer')
df_subset['weight'] = df_subset['weight'].fillna('Unknown Weight')
display(df_subset.isnull().sum())
```
